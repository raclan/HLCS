# Player Stats

```python
# return row of stats for given player

def get_stats(player, champion = False):
    
    if champion:
        ps_gr = raw_ps.loc[raw_ps['Champion'] == player].join(gr.set_index(['Series #', 'Game #']), on=['Series #', 'Game #'])
    else:
        ps_gr = raw_ps.loc[raw_ps['Player'] == player].join(gr.set_index(['Series #', 'Game #']), on=['Series #', 'Game #'])

    wins = np.count_nonzero((ps_gr['Blue Red'] == ps_gr['Winning Side']))

    losses = np.count_nonzero((ps_gr['Blue Red'] == ps_gr['Winning Side']) == 0)

    win_p = round(wins / (wins + losses), 3)

    kills = sum(ps_gr['Kills'])

    deaths = sum(ps_gr['Deaths'])

    assists = sum(ps_gr['Assists'])

    if deaths == 0:
        kda = kills + assists
    else:
        kda = round((kills + assists) / deaths, 3)

    cs = sum(ps_gr['CS'])

    vision_score = sum(ps_gr['Vision Score'])

    avg_vision_score = round(vision_score / (wins + losses), 3)

    fantasy_score = sum(ps_gr['Fantasy Score'])

    avg_fantasy_score = round(fantasy_score / (wins + losses), 3)

    total_kills = sum(ps_gr.loc[ps_gr['Blue Red'] == ps_gr['Winning Side']]['Winning Team Kills']) + sum(ps_gr.loc[(ps_gr['Blue Red'] == ps_gr['Winning Side']) == 0]['Losing Team Kills'])

    kp = round((kills + assists) / total_kills, 3)

    minutes = sum(ps_gr['Length'])

    csmin = round(sum(ps_gr['CS']) / minutes, 3)

    avg_kills = round(kills / (wins + losses), 3)

    avg_deaths = round(deaths / (wins + losses), 3)

    avg_assists = round(assists / (wins + losses), 3)

    kpm = round(kills / minutes, 3)

    dpm = round(deaths / minutes, 3)

    apm = round(assists / minutes, 3)

    damage_dealt = sum(ps_gr['Damage Dealt'])

    ddpm = round(damage_dealt / minutes, 3)

    return [player, wins, losses, win_p, kills, deaths, assists, kda, cs, vision_score, avg_vision_score, fantasy_score, avg_fantasy_score, total_kills, kp, minutes, csmin, avg_kills, avg_deaths, avg_assists, kpm, dpm, apm, damage_dealt, ddpm]
```

```python
final_ps = pd.DataFrame(columns=['Player', 'Wins', 'Losses', 'Win Percentage', 'Kills', 'Deaths', 'Assists', 'KDA', 'CS', 'Vision Score', 'Average Vision Score', 'Fantasy Score', 'Average Score' , 'Total Kills of Team' , 'Kill Participation' , 'Minutes Played' , 'CS/Min' , 'Average Kills', 'Average Deaths', 'Average Assists', 'Kills/Minute', 'Deaths/Minute', 'Assists/Minute', 'Damage Dealt', 'DPM'])
```

```python
for p in raw_ps['Player'].unique():
    final_ps.loc[len(final_ps)] = get_stats(p)
```

```python
final_ps = final_ps.sort_values(by=['Average Score', 'Win Percentage', 'Wins'], ascending=False).reset_index(drop=True)
final_ps
```

```python
# ryan_roles = raw_ps[raw_ps['Player'] == 'Ryan']['Role'].value_counts()

# ryan_top = 0
# ryan_jungle = 0
# ryan_mid = 0
# ryan_bottom = 0
# ryan_support = 0

# for i in range(len(ryan_roles)):
#     match ryan_roles.index[i]:
#         case 'Top':
#             ryan_top = ryan_roles[i]
#         case 'Jungle':
#             ryan_jungle = ryan_roles[i]
#         case 'Mid':
#             ryan_mid = ryan_roles[i]
#         case 'Bottom':
#             ryan_bottom = ryan_roles[i]
#         case 'Support':
#             ryan_support = ryan_roles[i]

# ryan_flex_score = 1 - np.std([ryan_top, ryan_jungle, ryan_mid, ryan_bottom, ryan_support] / (ryan_top + ryan_jungle + ryan_mid + ryan_bottom + ryan_support), ddof=1)
# ryan_flex_score
```

```python
def get_roles(player):
    roles = raw_ps[raw_ps['Player'] == player]['Role'].value_counts()

    top = 0
    jungle = 0
    mid = 0
    bottom = 0
    support = 0

    for i in range(len(roles)):
        match roles.index[i]:
            case 'Top':
                top = roles[i]
            case 'Jungle':
                jungle = roles[i]
            case 'Mid':
                mid = roles[i]
            case 'Bottom':
                bottom = roles[i]
            case 'Support':
                support = roles[i]

    flex_score = 1 - np.std([top, jungle, mid, bottom, support] / (top + jungle + mid + bottom + support), ddof=4)

    return [player, top, jungle, mid, bottom, support, flex_score]
```

```python
flex_scores = pd.DataFrame(columns=['Player', 'Top', 'Jungle', 'Mid', 'Bottom', 'Support', 'Flex Score'])

for p in raw_ps['Player'].unique():
    flex_scores.loc[len(flex_scores)] = get_roles(p)
```

```python
flex_scores = flex_scores.sort_values(by='Flex Score', ascending=False).reset_index(drop=True)
flex_scores
```

## All-Pro

```python
def get_role_scores(player):

    scores = [player]

    for r in ['Top', 'Jungle', 'Mid', 'Bottom', 'Support']:
        scores.append(np.sum(raw_ps[raw_ps['Player'] == player][raw_ps['Role'] == r]['Fantasy Score']))

    return scores
```

```python
role_scores = pd.DataFrame(columns=['Player', 'Top', 'Jungle', 'Mid', 'Bottom', 'Support'])

for p in raw_ps['Player'].unique():
    role_scores.loc[len(role_scores)] = get_role_scores(p)
```

```python
def get_player_champion_stats(player, champion):
    games = len(raw_ps[raw_ps['Player'] == player][raw_ps['Champion'] == champion])
    average_fantasy_score = round(np.mean(raw_ps[raw_ps['Player'] == player][raw_ps['Champion'] == champion]['Fantasy Score']), 3)
    variance = round(np.std(raw_ps[raw_ps['Player'] == player][raw_ps['Champion'] == champion]['Fantasy Score']), 3)

    return [player, champion, games, average_fantasy_score, variance]
```

```python
player_champion_stats = pd.DataFrame(columns=['Player', 'Champion', 'Games Played', 'Average Fantasy Score', 'Standard Deviation'])

for p in np.unique(raw_ps['Player']):
    for c in np.unique(raw_ps[raw_ps['Player'] == p]['Champion']):
        player_champion_stats.loc[len(player_champion_stats)] = get_player_champion_stats(p, c)
```

```python
def champ_stats(champion, games):
    rd_gr = rd.loc[rd['Champion'] == champion].join(gr.set_index(['Series #', 'Game #']), on=['Series #', 'Game #'])

    picks = np.count_nonzero(rd_gr['Pick Ban'] == 'Pick')

    bans = np.count_nonzero(rd_gr['Pick Ban'] == 'Ban')

    picked = rd_gr.loc[rd_gr['Pick Ban'] == 'Pick']
    wins = np.count_nonzero((picked['Blue Red'] == picked['Winning Side']))

    presence = round((picks + bans) / games, 3)

    pickrate = round(picks / games, 3)

    banrate = round(bans / games, 3)

    if len(picked) == 0:
        winrate = 0
    else:
        winrate = round(wins / len(picked), 3)

    return [champion, picks, bans, wins, presence, pickrate, banrate, winrate]
```

```python
champ_pb = pd.DataFrame(columns=['Champion', 'Picks', 'Bans', 'Wins', 'Presence', 'Pickrate', 'Banrate', 'Winrate'])
                        
games = len(rd.filter(['Series #', 'Game #']).drop_duplicates().dropna())

for c in rd['Champion'].unique():
    champ_pb.loc[len(champ_pb)] = champ_stats(c, games)
```

```python
champ_pb = champ_pb.sort_values('Champion').dropna().reset_index(drop=True)
champ_pb
```
